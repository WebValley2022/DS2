```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
frame1 = pd.read_csv('../exports/appa1.csv')
frame1.Time = pd.to_datetime(frame1.Time)
appa_columns = frame1.columns.to_numpy()[1:]
```

```python
fig = plt.figure(figsize=(25, 120))
fig.suptitle('Appa 1 Params')
outer = gridspec.GridSpec(len(appa_columns), 2, wspace=0.2, hspace=0.9)
for index, column in enumerate(appa_columns):
    ax = plt.Subplot(fig, outer[index])
    ax.scatter(frame1.Time, frame1[column], color='red', alpha=0.4, label=column)
    ax.set_ylabel(column)
    ax.set_xlabel('mean_test_score')
    ax.legend()
    fig.add_subplot(ax)
fig.show()
```
