```python
#Change the path to the file

prova2 = pd.read_csv("data/Ground_Truth.csv", encoding="ISO-8859-1")
prova2 = prova2.drop("Unite",axis = 1)
prova2 = prova2.dropna()
supp = prova2.loc[prova2.Inquinante == "Inquinante"].index
prova2 = prova2.drop(supp,axis = 0).reset_index()
#locale.setlocale(locale.LC_TIME, 'en_IN')
prova2.Data = pd.to_datetime(prova2.Data , format = "%Y-%m-%d")
for i, row in tqdm(prova2.iterrows(), total=prova2.shape[0]):
    prova2.at[i,"Data"] = row["Data"]  + pd.DateOffset(hours=int(row["Ora"]))
prova2.rename(columns = {'Data':'Time'}, inplace = True)

loc = ["Parco S. Chiara", "Via Bolzano"]
Gas = ["PM10", "Ossido di Carbonio", "Biossido di Azoto", "Biossido Zolfo", "Ozono"]
```

```python
#Divide the Dataset in a readable way

supp = prova2.drop(["index","Ora"], axis = 1)

supp = supp.set_index("Time")

APPA1 = pd.DataFrame()
APPA2 = pd.DataFrame()
for i in loc:
    for j in Gas:
        supp2 = supp.loc[supp.Stazione == i]
        supp3 = supp2.loc[supp2.Inquinante == j]
        if supp3.empty:
            continue
        
        prova3 = pd.DataFrame(data = supp3.Valore.values, columns = [j], index = supp3.index)
        if i == loc[0]:
            if APPA1.empty:
                APPA1 = prova3.reset_index()
            else:
                #APPA1 = jp.pd.concat([prova.reset_index(),APPA1], axis = 1)
                APPA1 = APPA1.merge(prova3.reset_index(), how = "outer")
        else:
            if APPA2.empty:
                APPA2 = prova3.reset_index()
            else:
                #APPA1 = jp.pd.concat([prova.reset_index(),APPA1], axis = 1)
                APPA2 = APPA2.merge(prova3.reset_index(), how = "outer")
```

```python
APPA2.dropna(how = "any",inplace=True)
APPA2.loc[(APPA2.Time > pd.to_datetime(1635634800, unit='s')) & (APPA2.Time < pd.to_datetime(1648422000, unit='s') ),"Time"] = APPA2.loc[(APPA2.Time > pd.to_datetime(1635634800, unit='s')) & (APPA2.Time < pd.to_datetime(1648422000, unit='s') )].Time - pd.Timedelta(1, unit="H") 
APPA2 = APPA2.drop_duplicates()
APPA2 = APPA2.set_index("Time")
```

```python
APPA1.dropna(how = "any",inplace=True)
APPA1.loc[(APPA1.Time > pd.to_datetime(1635634800, unit='s')) & (APPA1.Time < pd.to_datetime(1648422000, unit='s') ),"Time"] = APPA1.loc[(APPA1.Time > pd.to_datetime(1635634800, unit='s')) & (APPA1.Time < pd.to_datetime(1648422000, unit='s') )].Time - pd.Timedelta(1, unit="H") 
APPA1 = APPA1.drop_duplicates()
APPA1 = APPA1.set_index("Time")
```
